```python
import requests
import json
import uuid

URL = "http://localhost:8000/predict_batch"

# T·∫°o batch comment m·∫´u ƒë·ªÉ test
payload = {
    "batch": [
        {
            "id": str(uuid.uuid4()),
            "text": "m√†y ƒë√∫ng l√† ƒë·ªì_ngu, kh√¥ng bi·∫øt suy_nghƒ© g√¨ c·∫£"
        },
        {
            "id": str(uuid.uuid4()),
            "text": "s·∫£n_ph·∫©m n√†y d√πng kh√° t·ªët, m√¨nh th·∫•y ·ªïn"
        },
        {
            "id": str(uuid.uuid4()),
            "text": "b·ªçn n√†y to√†n l·ª´a_ƒë·∫£o, kh√¥ng n√™n tin"
        }
    ]
}

print("üì§ Sending request...")
resp = requests.post(URL, json=payload, timeout=30)

print("Status code:", resp.status_code)

if resp.status_code == 200:
    data = resp.json()
    print("\nüì• API Response:")
    print(json.dumps(data, indent=2, ensure_ascii=False))
else:
    print("‚ùå Error response:")
    print(resp.text)
```
